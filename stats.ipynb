# Работа с различными распределениями

### Равномерное распределение
Равномерное распределение — это непрерывное распределение, при котором вероятность появления любого значения на заданном интервале одинакова. Если случайная величина $$X$$ равномерна на интервале $$[a, b]$$, то вероятность $$P(a \leq X \leq b) = \frac{1}{b-a}$$ для всех $$x$$ из $$[a, b]$$.

```python
uniform_data = np.random.uniform(low=0.0, high=1.0, size=1000)
```

```python
uniform_data
```

```python
sns.histplot(uniform_data)
```

```python
sns.boxplot(uniform_data)
```

### Нормальное распределение
Нормальное распределение, или гауссовское, — это симметричное распределение в виде "колокола", определяемое математическим ожиданием (средним значением, $$\mu$$) и стандартным отклонением ($$\sigma$$). Функция плотности вероятности имеет вид:
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp \left( -\frac{(x - \mu)^2}{2 \sigma^2} \right)
$$
Используется как базовая модель в статистике и машинном обучении.

```python
normal_data = np.random.normal(loc=0.0, scale=1.0, size=1000)
```

```python
sns.histplot(normal_data)
```

```python
sns.boxplot(normal_data)
```

### Биномиальное распределение
Биномиальное распределение описывает число успехов в серии из $$n$$ независимых испытаний с вероятностью успеха $$p$$ в каждом. Вероятность получить $$k$$ успехов вычисляется по формуле:
$$
P(X=k) = C_n^k p^k (1-p)^{n-k}
$$
где $$C_n^k$$ — биномиальный коэффициент. Пример: количество орлов при десятикратном подбрасывании монеты.

```python
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
```

```python
binomial_data
```

```python
sns.histplot(binomial_data)
```

```python
sns.boxplot(binomial_data)
```

### Экспоненциальное распределение
Экспоненциальное распределение — это непрерывное распределение для времени между событиями в пуассоновском процессе. Функция плотности:
$$
f(x; \lambda) = \lambda e^{-\lambda x}
$$
для $$x \geq 0$$, где $$\lambda$$ — параметр интенсивности. Используется, например, для описания времени до следующего события (отказа устройства, звонка и т.д.).

```python
exponential_data = np.random.exponential(scale=1.0, size=1000)
```

```python
sns.boxplot(exponential_data)
```

### Пуассоновское распределение
Пуассоновское распределение моделирует количество событий, произошедших за фиксированный промежуток времени или пространства, при средней интенсивности $$\lambda$$. Формула вероятности:
$$
P(X=k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$
Для дискретных моделей событий в больших объемах ("сколько клиентов пришло в магазин за 1 час").

```python
poisson_data = np.random.poisson(lam=3.0, size=1000)
```

### Бета-распределение
Бета-распределение — это непрерывное распределение на интервале [1] с двумя параметрами $$a$$ и $$b$$, используется для моделирования вероятностей и пропорций. Плотность вероятности:
$$
f(x; a, b) = \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} x^{a-1}(1-x)^{b-1}
$$
Главным образом используется в байесовской статистике.

```python
beta_data = np.random.beta(a=2.0, b=5.0, size=1000)
```

```python
sns.boxplot(beta_data)
```

# Вычисление CDF/PMF/PDF

Для вычисления функции плотности вероятности (PDF) и функции распределения (CDF) на Python обычно используют библиотеку SciPy, которая предоставляет готовые функции для многих распространённых распределений.

### Пример для нормального распределения с SciPy

```python
mu, sigma = 0, 1

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

pdf = stats.norm.pdf(x, mu, sigma)

cdf = stats.norm.cdf(x, mu, sigma)

plt.plot(x, pdf, label='Плотность вероятности (PDF)')
plt.plot(x, cdf, label='Функция распределения (CDF)')
plt.legend()
plt.title('Нормальное распределение')
plt.show()

# `norm.pdf(x, mu, sigma)` — вычисляет значение плотности вероятности в точках массива `x`.
# `norm.cdf(x, mu, sigma)` — вычисляет значение функции распределения в точках `x` (накопленная вероятность).
```

### Общая схема для других распределений

Большинство классических распределений в `scipy.stats` имеют методы `.pdf()` и `.cdf()`. Например:

| Распределение | Импорт и объект | Пример вызова PDF | Пример вызова CDF |
|---------------|-----------------|------------------|-------------------|
| Нормальное | `from scipy.stats import norm` | `norm.pdf(x, loc, scale)` | `norm.cdf(x, loc, scale)` |
| Биномиальное | `from scipy.stats import binom` | `binom.pmf(k, n, p)` (pmf — для дискретных) | `binom.cdf(k, n, p)` |
| Пуассона | `from scipy.stats import poisson` | `poisson.pmf(k, mu)` | `poisson.cdf(k, mu)` |
| Экспоненциальное | `from scipy.stats import expon` | `expon.pdf(x, scale=1/lambda)` | `expon.cdf(x, scale=1/lambda)` |

***

### Объяснение

- PDF (функция плотности вероятности) показывает плотность вероятности на конкретном значении.
- CDF (функция распределения) показывает накопленную вероятность, что случайная величина меньше или равна заданному значению.
- Для дискретных распределений используется PMF (функция вероятности), которая аналогична PDF, но для дискретных событий.

# Решение задач на Python

### Задача
Дана выборка из некоторого нормального распределения (например, измерения роста группы людей). Нужно определить 95% доверительный интервал для истинного среднего значения.

```python
np.random.seed(42)
sample = np.random.normal(loc=170, scale=10, size=50)

mean_sample = np.mean(sample)
std_sample = np.std(sample, ddof=1)
n = len(sample)

alpha = 0.05

t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

margin_error = t_crit * (std_sample / np.sqrt(n))

ci_lower = mean_sample - margin_error
ci_upper = mean_sample + margin_error

print(f"Выборочное среднее: {mean_sample:.2f}")
print(f"95% доверительный интервал: ({ci_lower:.2f}, {ci_upper:.2f})")
```

### Объяснение

- Генерируем выборку из нормального распределения с известными параметрами.
- Считаем выборочное среднее и стандартное отклонение.
- По уровню надежности выбираем критическое значение статистики t с помощью `stats.t.ppf`.
- Вычисляем доверительный интервал по формуле:
  $$
  \bar{x} \pm t_{\alpha/2, n-1} \times \frac{s}{\sqrt{n}}
  $$
- Выводим результат.

# АБ

### Задача
Есть две группы (A и B), каждая с известным числом пользователей и числом совершивших целевое действие (например, покупку). Нужно проверить, есть ли статистически значимая разница в конверсии между группами.

```python
users = np.array([1000, 1200])
conversions = np.array([200, 270])

table = np.array([
    [conversions[0], users[0] - conversions[0]],
    [conversions[1], users[1] - conversions[1]]
])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(table)

print(f"Chi2 статистика: {chi2_stat:.4f}")
print(f"p-значение: {p_value:.4f}")

if p_value < 0.05:
    print("Разница статистически значима, группы отличаются по конверсии")
else:
    print("Разница не является статистически значимой")
```

### Объяснение
- Формируется таблица с наблюдаемыми значениями «конверсия/отсутствие конверсии» для групп A и B.
- Применяется тест независимости χ², который проверяет гипотезу о равенстве долей конверсии.
- Если p-значение меньше выбранного уровня значимости (обычно 0.05), то отклоняется гипотеза о равенстве долей, значит группы различаются по эффективности.

# АБ (Бутстрап)

### Задача
Есть две группы с показателями конверсии. Нужно оценить разницу конверсий и её доверительный интервал с помощью бутстрапа.

```python
users_A, conv_A = 1000, 200
users_B, conv_B = 1200, 270

data_A = np.concatenate([np.ones(conv_A), np.zeros(users_A - conv_A)])
data_B = np.concatenate([np.ones(conv_B), np.zeros(users_B - conv_B)])


def bootstrap_diff(data1, data2, n_bootstrap=10000):
    diffs = []
    for _ in range(n_bootstrap):
        sample1 = np.random.choice(data1, size=len(data1), replace=True)
        sample2 = np.random.choice(data2, size=len(data2), replace=True)
        diff = np.mean(sample2) - np.mean(sample1)
        diffs.append(diff)
    return np.array(diffs)


bootstrap_diffs = bootstrap_diff(data_A, data_B)

ci_lower = np.percentile(bootstrap_diffs, 2.5)
ci_upper = np.percentile(bootstrap_diffs, 97.5)

p_value = np.mean(bootstrap_diffs <= 0)

print(f"Средняя разница конверсии (B - A): {np.mean(bootstrap_diffs):.4f}")
print(f"95% доверительный интервал: ({ci_lower:.4f}, {ci_upper:.4f})")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("Разница статистически значима")
else:
    print("Разница не является статистически значимой")
```

```python
sns.histplot(bootstrap_diffs)
```

### Объяснение
- Сначала моделируем данные каждой группы как массивы из 1 и 0 (конверсия и отсутствие).
- Метод бутстрапа случайными многократными выборками с возвращением оценивает распределение разницы конверсий.
- Доверительный интервал строится как перцентили эмпирического распределения разницы.
- p-value показывает вероятность наблюдать равную или меньшую разницу при отсутствии эффекта.
