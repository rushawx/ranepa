```python
from time import time
```

```python
def foo_timer(foo):
    def inner(*args, **kwargs):
        start = time()
        output = foo(*args, **kwargs)
        end = time()
        print(f"{foo.__name__}: executed for {round(end - start, 4)} seconds and returned {output}")
    return inner
```

```python
def time_consuming_function(num):
    return sum(range(num))
```

```python
time_consuming_function(10**9)
```

```python
@foo_timer
def time_consuming_function(num):
    return sum(range(num))
```

```python
time_consuming_function(10**9)
```

```python
def foo_timer(n):
    def outer(foo):
        def inner(*args, **kwargs):
            data = []
            for i in range(n):
                start = time()
                output = foo(*args, **kwargs)
                end = time()
                print(f"{i} iteration. {foo.__name__}: executed for {round(end - start, 4)} seconds and returned {output}")
                data.append(end - start)
            print(f"{foo.__name__} on average took {round(sum(data) / len(data), 4)} seconds")
        return inner
    return outer
```

```python
@foo_timer(3)
def time_consuming_function(num):
    return sum(range(num))
```

```python
time_consuming_function(10**9)
```

# Функциональное программирование

```python
def avg_procedure(num):
    sum_value = 0
    count = 0
    for i in range(num):
        sum_value += i
        count += 1
    return sum_value / count
```

```python
avg_procedure(10)
```

```python
def avg_functional(num):
    return sum(range(num)) / len(range(num))
```

```python
avg_functional(10)
```

```python
avg_lambda = lambda x: sum(range(x)) / len(range(x))
```

```python
avg_lambda(10)
```

## ООП - Инкапсуляция

```python
class MyMisterousCat(Cat):
    def __init__(self, name):
        super().__init__(
            name=name,
        )
        self.__private_var__ = 42
        self._protected_var_ = 7
        self.public_var = 3
    
    def __my_private_method__(self):
        print(self.__private_var__)
    
    def _my_protected_method_(self):
        print(self._private_var_)

    def my_public_method(self):
        print(self.__private_var__)
        print(self.public_var)
    
    def make_sound(self):
        print(self.sound.upper())
```

```python
some_cat = MyMisterousCat(name="Surprise")
```

```python
some_cat.__private_var__
```

```python
some_cat.__my_private_method__()
```

```python
some_cat.my_public_method()
```

```python
some_cat._protected_var_
```

```python
some_cat.public_var
```

# Итераторы

```python
class MyIterator:
    def __init__(self, threshold):
        self.threshold = threshold
        self.current = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current < self.threshold:
            self.current += 1
            return self.current
        raise StopIteration
```

```python
my_iter = MyIterator(5)
```

```python
next(my_iter)
```

```python
next(my_iter)
```

```python
next(my_iter)
```

```python
next(my_iter)
```

```python
next(my_iter)
```

```python
try:
    next(my_iter)
except StopIteration:
    print(f"caught exception")
```

# Контекстные менеджеры

```python
with open("example.txt", "w") as f:
    f.write("hello")
```

```python
with open("example.txt", "r") as f:
    for line in f.readlines():
        print(line)
```

```python
class CustomOpen:
    def __init__(self, path):
        self.path = path
        self.file = None
    
    def __enter__(self):
        self.file = open(self.path, "r")
        print(f"just opened file by path - {self.path}")
        return self.file
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.file:
            self.file.close()
        print(f"just closed file by path - {self.path}")
```

```python
with CustomOpen("example.txt") as f:
    print(f.read())
```
