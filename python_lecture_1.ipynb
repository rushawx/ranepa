# Управление потоком выполнения программ

## for

```python
for i in range(10):
    print(i)
```

```python
for i in range(5, 10):
    print(i)
```

```python
for i in range(5, 10, 2):
    print(i)
```

```python
my_list
```

```python
for i in my_list:
    print(i)
```

```python
for i in my_tuple:
    print(i)
```

```python
for i in my_dict:
    print(i)
```

```python
for value in my_dict.values():
    print(value)
```

```python
for key, value in my_dict.items():
    print(f"{key}: {value}")
```

## while

```python
check = True
counter = 0

while check:
    print(counter)
    counter += 1
    if counter > 10:
        check = False
```

## if

```python
if 4 % 2 == 0:
    print("4 is an even number")
```

```python
for i in range(10):
    if i % 2 == 0:
        print(f"{i}: even")
    else:
        print(f"{i}: odd")
```

```python
for i in range(10):
    if i % 2 == 0:
        print(f"{i}: even")
    elif i % 3 == 0:
        print(f"{i}: is divided by 3")
    else:
        print(f"{i}: odd")
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```
